Destarte, trabalharei com o dataset order_payments.

Agora, estudarei quais pagamentos são feitos com mais frequẽncia, o valor médio
dos pagamentos e seu desvio padrão. Usarei gráficos também.

```python
order_payments.info()
```

```python
# Frequência dos pagamentos
order_payments['payment_type'].value_counts()
```

```python
# Remoção dos dados não definidos
order_payments = order_payments[order_payments['payment_type'] != 'not_defined']
```

```python
# Média dos pagamentos por tipo de pagamento
order_payments.groupby('payment_type')['payment_value'].mean()
```

```python
# Desvio padrão dos pagamentos por tipo de pagamento
order_payments.groupby('payment_type')['payment_value'].std()
```

```python
# Análise gráfica

# Tomar a média e o desvio padrão
media = order_payments.groupby('payment_type')['payment_value'].mean()
desvio_padrao = order_payments.groupby('payment_type')['payment_value'].std()

# Plotar os valores de 'payment_value' por 'payment_type', com a média e o desvio padrão plotados como linhas
plt.errorbar(media.index, media, yerr=desvio_padrao, capsize=5)

# Adicionar legenda e eixos
plt.xlabel('Tipo de pagamento')
plt.ylabel('Valor da média')
plt.grid(True)

# Exibir o gráfico
plt.show()
```

6. Na tabela de produtos você pode criar uma coluna categórica com valores
como “pequeno”, “médio”, e “grande” para cada produto, a depender de
seus valores de height e width.

```python
# Calcular a área de cada produto
products['area'] = products['product_height_cm'] * products['product_length_cm']

# Calcular a média das áreas
area_mean = products['area'].mean()
print(f"A área média vale {area_mean} cm2.")

# Classificar cada produto como "small", "medium", ou "large"
# Será tomada uma margem de +/- 10% da área média para classificação
products['size'] = products.apply(lambda row: 'small' if row['area'] < 0.9 * area_mean else
                                            ('medium' if row['area'] >= 0.9 * area_mean and 
                                             row['area'] <= 1.1 * area_mean else 
                                             'large'), axis=1)
```
